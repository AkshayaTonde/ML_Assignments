```python
warnings.filterwarnings('ignore') # adding entry into the specifications of the warnings filter.
Health_Scrng = pd.read_csv("new.csv")
Health_Scrng['BMICat'] = Health_Scrng['BMICat'].map({'Normal':0,'Obese':1,'Under Weight':2,'Over Weight':3})

df = pd.DataFrame(Health_Scrng,columns= ['AgeinYr','height','weight','BMICat']) # Calling DataFrame constructor on list

X= Health_Scrng.drop(['age','id','ap_hi','ap_lo','AgeGroup','BMICat'],axis=1) # Assignment of x variable (Independent variable which is also called as responser)
y = df['BMICat'] # Assignment of y variable (Dependent variable which is also called as predictor)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=3) # splitting data arrays into two subsets: for training data and for testing data.


```

**Confusion Matrix**

```python
logistic_regression= LogisticRegression(multi_class='multinomial', solver='lbfgs') # LogisticRegression class can be configured for multinomial logistic regression
logistic_regression.fit(X_train,y_train)  # used for best representative function for the the data points
y_pred=logistic_regression.predict(X_test) # Prediction variable assignment 

confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted']) # method is used to compute a simple cross-tabulation of two (or more) factors
sn.heatmap(confusion_matrix, annot=True) # a plot Confusion matrix.

print('Accuracy: ',metrics.accuracy_score(y_test, y_pred)) # Printing accuracy of dataset
print("Precision:",metrics.precision_score(y_test, y_pred,average='weighted'))
print("Recall:",metrics.recall_score(y_test, y_pred,average='weighted'))
plt.show()
```
